random forest - MULTICLASS CLASSFICATION

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
sns.set()
```

```python
data = pd.read_csv('Digital recognition.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
x = data.drop(['target'], axis=1)

# df.drop(['a', 'b'], axis=1)
```

```python
x.head()
```

```python
x_scaled = preprocessing.scale(x)
```

```python
x_scaled
```

```python
#y = data.iloc[:, 65]
y = data['target']
```

```python
y.describe()
```

```python
y.unique()
```

```python
y.value_counts()
```

SELECTING 20% FOR TESTING AND 80% FOR TRAINING

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, c =  train_test_split(x_scaled, y, test_size = 0.2)
```

NOW RUNNING THE MODEL RANDOM FOREST

```python
model = RandomForestClassifier()
```

```python
r = model.fit(x_train, y_train)
```

```python
r
```

```python
model.score(x_test, y_test)
```

```python
p = model.predict(x_test)
```

```python
p.shape
```

```python
y_test.shape
```

```python
compare = pd.DataFrame()
compare['y_test'] = y_test
compare['predicting'] = p
compare['Matching'] = np.where(y_test == p, 1, 0)
compare.describe()
```

```python
counts = (compare['Matching']==1).value_counts().to_dict()
counts
```

```python
accuracy = counts[True]/(counts[True] + counts[False])
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(p, y_test)
cm
```

```python
#MAKING A DATAFRAME FOR THE CONFUSION MATRIX

cm_df = pd.DataFrame(cm)
cm_df.columns = ['P 0','P 1','P 2','P 3','P 4','P 5','P 6','P 7','P 8','P 9']
cm_df = cm_df.rename(index={0: 'Actual 0',1:'Actual 1',2:'Actual 2',3:'Actual 3',4:'Actual 4',5:'Actual 5',6:'Actual 6',7:'Actual 7',8:'Actual 8',9:'Actual 9'})
cm_df
```
