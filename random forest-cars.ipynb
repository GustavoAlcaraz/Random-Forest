dropping model because there are many models (267) to convert in a dummy variable

BECAUSE THE MODEL IS RANDOM FOREST WE NEED TO CATEGORIZE PRICE

```python
bins = [0, 1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 900000]
names = ['<1000', '1k-5k', '5k-10k', '10k-20k', '20k-30k','30k-40k', '40k-50k','50k-60k','60k-70k','70k-80k','80k-90k''90k+']
```

```python
d = dict(enumerate(names, 1))
```

```python
raw_data['Price Range'] = np.vectorize(d.get)(np.digitize(raw_data['Price'], bins))
```

```python
raw_data = raw_data.drop(['Price'], axis = 1)
```

```python
raw_data.head()
```

```python
raw_data.columns
```

```python
raw_data = raw_data[['Price Range', 'Brand', 'Body', 'Mileage', 'EngineV', 'Engine Type', 'Registration',
       'Year']]
```

```python
raw_data.head()
```

DUMMY VARIABLES

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Brand = LabelEncoder()
le_Body = LabelEncoder()
le_Engine_Type = LabelEncoder()
le_Registration = LabelEncoder()
le_Price_Range = LabelEncoder()
```

```python
raw_data['Brand_n'] = le_Brand.fit_transform(raw_data['Brand'])
raw_data['Body_n'] = le_Body.fit_transform(raw_data['Body'])
raw_data['Engine_Type_n'] = le_Engine_Type.fit_transform(raw_data['Engine Type'])
raw_data['Registration_n'] = le_Registration.fit_transform(raw_data['Registration'])
raw_data['Price_Range_n'] = le_Price_Range.fit_transform(raw_data['Price Range'])
```

```python
raw_data.head()
```

```python
raw_data = raw_data.drop(['Brand', 'Body', 'Engine Type', 'Registration', 'Price Range'], axis = 1)
```

```python
raw_data.head() 
```

```python
raw_data.reset_index(drop = True)
```

```python
data_clean = raw_data.copy()
```

```python
data_clean.head()
```

```python
data_clean.describe(include = 'all')
```

SELECTING INPUTS AND TARGET

```python
x = data_clean.drop(['Price_Range_n'], axis = 1)
```

```python
y = data_clean['Price_Range_n']
```

```python
y.head()
```

SELECTING DATA TO BE TESTED AND TRAINING

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,  y_test = train_test_split(x,y, test_size = 0.2, random_state=42)
```

NOW RUNNING THE MODEL RANDOM FOREST

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train )
```

```python
p = model.predict(x_test)
```

```python
model.score(x_test, y_test)
```

```python
compare = pd.DataFrame()
compare['y_test'] = y_test
compare['predicting'] = p
compare['Matching'] = np.where(y_test == p, 1, 0)
compare.describe()
```

```python
counts = (compare['Matching']==1).value_counts().to_dict()
counts
```

```python
accuracy = counts[True]/(counts[True] + counts[False])
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(p, y_test)
cm
```
